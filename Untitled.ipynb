#Dataset churn_modelling

```python
import pandas as pd
```

```python
dataset = pd.read_csv("Churn_Modelling.csv") 
```

```python
dataset.head(5)
```

```python
dataset.columns
```

```python
Y = dataset['Exited']
```

```python
Y.head(5)
```

```python
X_raw = dataset[['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard',
       'IsActiveMember', 'EstimatedSalary']]
```

```python
X_raw.head(5)
```

```python
geo = pd.get_dummies(dataset['Geography'], drop_first = True)
```

```python
geo
```

```python
gender = pd.get_dummies(dataset['Gender'], drop_first = True)
```

```python
gender
```

```python
X = pd.concat([X_raw,geo,gender], axis = 1)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split( X, Y,test_size = 0.20, random_state = 40)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_test.shape
```

```python
Y_train.shape
```

```python
from keras.models import Sequential
```

```python
model = Sequential() #creating empty model
```

```python
from keras.layers import Dense
```

```python
#add first layer : input layer to thge model , 8-neurons, activation function is relu, input_dimentions 11columns
model.add(Dense(units=8, activation = 'relu', input_dim=11))
```

```python
#add second layer: hidden layer to the model , 6-nuerons
model.add(Dense(units=6, activation = 'relu'))
```

```python
#add third layer: hidden layer to the model , 6-nuerons
model.add(Dense(units=6, activation = 'relu'))
```

```python
#add last layer: hidden layer to the model , 1-nuerons, sigmoid = 1 or 0
model.add(Dense(units=1, activation = 'sigmoid'))
```

```python
model.get_config()
```

```python
model.summary()
```

```python
from keras.optimizers import Adam
```

```python
model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.000001))
```

```python
model.get_weights()
```

```python
model.fit(X_train, Y_train, epochs=100)
```

```python
#model loss history
loss_history = model.history.history
```

```python
loss_history
```

```python
loss= pd.DataFrame(loss_history)
loss
```

```python
loss.plot()
```
